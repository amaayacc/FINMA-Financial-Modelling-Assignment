```python
# Load Data
import yfinance as yf
import matplotlib.pyplot as plt

ticker = "AAPL"   # choose any ticker e.g. AAPL, MSFT, AMD

data = yf.download(ticker, start="2015-01-01")
print(data)

# Display the visualisation of the data
plt.plot(data['Close'])
plt.title(f"{ticker} Closing Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python
import pandas as pd
prices = pd.Series([100, 102, 101, 105, 103, 107])

# Calculate a simple moving average
ma = prices.rolling(window=3).mean()

# Generate signal: 1 if price > moving average, else 0
signal = (prices > ma).astype(int)

print(signal)

print("signal created vinuki")
```
